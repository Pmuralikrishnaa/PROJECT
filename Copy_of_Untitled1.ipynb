```python
import random

# Constants
EMPTY = ' '
HUMAN_PLAYER = 'X'
AI_PLAYER = 'O'
Player = input("What’s your name?")

while True:
    try:
        age = int(input(Player + " What's your age? "))
        break
    except ValueError:
        print("Invalid input. Please enter a number")

print(20*' ',"   reference:    ")
print(20*' ','     |    |      ')
print(20*' ','  7  | 8  | 9    ')
print(20*' ',"-----+----+----- ")
print(20*' ',"     |    |      ")
print(20*' ',"  4  | 5  | 6    ")
print(20*' ',"-----+----+----- ")
print(20*' ',"     |    |      ")
print(20*' ',"  1  | 2  | 3    \n")

def print_board(board):
    print()
    print('                               reference:')
    print('     |    |     ',10*' ','     |    |   ',)
    print('  '+board[7]+'  | '+board[8]+'  | '+board[9]+'   ',10*' ','  7  | 8  | 9  ')
    print('-----+----+-----',10*' ',"-----+----+-----")
    print('     |    |     ',10*' ',"     |    |     ")
    print('  '+board[4]+'  | '+board[5]+'  | '+board[6]+'   ',10*' ',"  4  | 5  | 6   ")
    print('-----+----+-----',10*' ',"-----+----+-----")
    print('     |    |     ',10*' ',"     |    |      ")
    print('  '+board[1]+'  | '+board[2]+'  | '+board[3]+'   ',10*' ',"  1  | 2  | 3    \n\n")


def is_full(board):
    """Checks if the board is full."""
    for row in board:
        if EMPTY in row:
            return False
    return True


def get_empty_cells(board):
    """Returns a list of empty cells on the board."""
    empty_cells = []
    for i in range(1,10):
        for j in range(1,10):
            if board[i][j] == EMPTY:
                empty_cells.append((i, j))
    return empty_cells


def is_winner(board, player):
    winning_place = [[1,2,3],[4,5,6],[7,8,9],[1,4,7],[2,5,8],[3,6,9],[1,5,9],[3,5,7]]
    for win_place in winning_place:
        if board[win_place[0]] == board[win_place[1]] == board[win_place[2]] == player:
            return True


def evaluate(board):
    """Evaluates the current state of the board."""
    if is_winner(board, AI_PLAYER):
        return 1  # AI wins
    elif is_winner(board, HUMAN_PLAYER):
        return -1  # Human wins
    else:
        return 0  # Draw


def minimax(board, depth, maximizing_player):
    """Minimax algorithm with alpha-beta pruning."""
    if is_winner(board, AI_PLAYER):
        return 1
    elif is_winner(board, HUMAN_PLAYER):
        return -1
    elif is_full(board):
        return 0

    if maximizing_player:
        max_eval = float('-inf')
        for i, j in get_empty_cells(board):
            board[i][j] = AI_PLAYER
            eval = minimax(board, depth + 1, False)
            board[i][j] = EMPTY
            max_eval = max(max_eval, eval)
        return max_eval
    else:
        min_eval = float('inf')
        for i, j in get_empty_cells(board):
            board[i][j] = HUMAN_PLAYER
            eval = minimax(board, depth + 1, True)
            board[i][j] = EMPTY
            min_eval = min(min_eval, eval)
        return min_eval


def get_best_move(board, difficulty):
    """Returns the best move for the AI player using the minimax algorithm."""
    if difficulty == 'easy':
        # Random move for easy difficulty
        return random.choice(get_empty_cells(board))
    elif difficulty == 'medium':
        # Random move or optimal move for medium difficulty
        if random.randint(0, 100) < 80:
            return random.choice(get_empty_cells(board))
        else:
            best_eval = float('-inf')
            best_move = None
            for i, j in get_empty_cells(board):
                board[i][j] = AI_PLAYER
                eval = minimax(board, 0, False)
                board[i][j] = EMPTY

                if eval > best_eval:
                    best_eval = eval
                    best_move = (i, j)

            return best_move
    else:
        # Optimal move for hard difficulty
        best_eval = float('-inf')
        best_move = None
        for i, j in get_empty_cells(board):
            board[i][j] = AI_PLAYER
            eval = minimax(board, 0, False)
            board[i][j] = EMPTY

            if eval > best_eval:
                best_eval = eval
                best_move = (i, j)

        return best_move


def play():
    """Plays the Tic Tac Toe game."""
    board =['',' ',' ',' ',' ',' ',' ',' ',' ',' ']

    print("Tic Tac Toe - You are X, AI is O")
    print_board(board)

    difficulty =input(Player+"Choose difficulty (easy, medium, hard): ")

    while not is_full(board) and not is_winner(board, HUMAN_PLAYER):
        # Human's turn
        row = int(input("Enter the row (0-2): "))
        col = int(input("Enter the column (0-2): "))

        if board[row][col] != EMPTY:
            print("Invalid move. Try again.")
            continue

        board[row][col] = HUMAN_PLAYER
        print_board(board)

        if is_winner(board, HUMAN_PLAYER):
            print("You win!")
            break

        if is_full(board):
            print("It's a draw!")
            break

        # AI's turn
        print("AI's turn...")
        ai_row, ai_col = get_best_move(board, difficulty)
        board[ai_row][ai_col] = AI_PLAYER
        print_board(board)

        if is_winner(board, AI_PLAYER):
            print("AI wins!")
            break

        if is_full(board):
            print("It's a draw!")
            break


# Start the game
play()
```
